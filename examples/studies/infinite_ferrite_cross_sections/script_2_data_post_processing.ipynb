```python
import mph
import mphsweepkit as msk
```

```python
# Start the COMSOL client
client = mph.start()
```

```python
# Load the model
model = client.load('core_cross_section_solved.mph')
```

Load an already solved CascadedSweepModel

```python
csm = msk.CascadedSweepModel(model, 'Study on Cross-Sections', show_param_names=True)
```

Post-processing 

```python
# Load the customized post-processing expressions from a JSON file
post_processing_exprs = msk.load_post_processing_exprs("global_post_processing_expressions.json", print_info=True)
```

```python
csm.post_process_data(post_processing_exprs)
```

```python
display(csm.combined_data.head())
```

Save post-processing results

```python
csm.save_global_data()
```

Release Model

```python
client.remove(model)
client.clear()
client.disconnect()
```
