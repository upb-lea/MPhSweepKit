#### Permittivity Measurement at high frequencies

The simulated setup is an  idealized cuboid plate capactior


```python
%load_ext autoreload
%autoreload 2
```

```python
import mph
import mphsweepkit as msk
```

Start the Comsol Server and load the model

```python
# Start the COMSOL client
client = mph.start()
```

```python
# Load the model
model = client.load('cuboid_capacitor_initial.mph')
```

Create a CascadedSweepModel

```python
# Create a CascadedSweepModel object
csm = msk.CascadedSweepModel(model, 'Cuboid Capacitor Study', show_param_names=True)
csm.save_global_data()
```

Access the model tree in Comsol:

```python
csm.set_material_sweep(sweep_name='Material Sweep', 
                       material_names=['matsw.comp1.sw1'], 
                       material_values=[[1, 2, 3]], 
                       sweep_type="filled")

csm.input_data
```

Simulation

```python
csm.simulate()
```

Save the model with solution data:

```python
# Save to new file
model.save('cuboid_capacitor_solved.mph')
```

Release the Client

```python
client.remove(model)
client.clear()
client.disconnect()
```
